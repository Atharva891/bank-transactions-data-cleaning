```python
import pandas as pd
```

```python
df = pd.read_csv("raw_bank_transactions.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['transaction_id'] = pd.to_datetime(df['transaction_date'],errors='coerce',dayfirst=True)
```

```python
df['transaction_date']
```

```python
df
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
# TEXT CLEANING
df['transaction_type'] = df['transaction_type'].str.lower().str.strip()
```

```python
# FIXING NAME INCONSISTENCIES
df['transaction_type']=df['transaction_type'].replace({'withdraw':'withdrawal'})
```

```python
# CHECK UNIQUES VALUES
df['transaction_type'].unique()
```

```python
df
```

```python
# HANDLINIG MISSING VALUES
df['transaction_type'] = df['transaction_type'].fillna('unkown')
```

```python
# CONVERT NUMERIC VALUES
df['amount'] = pd.to_numeric(df['amount'],errors='coerce')
```

```python
df['amount']
```

```python
# HANDEL MISSING AMOUNTS
df=df.dropna(subset=['amount'])
```

```python
df
```

```python
#CLEANING CITY COL.
df.loc[:,'city']=(df['city'].str.strip().str.title())
```
